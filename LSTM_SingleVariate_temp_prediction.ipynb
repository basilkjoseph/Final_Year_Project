```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df=pd.read_csv(csv_path)
df
```

```python
df=df[5::6]
df.head()
```

```python
df.index=pd.to_datetime(df['Date Time'],format='%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
temp=df['T (degC)']
temp.plot()
```

```python
def df_to_X_y(df, window_size=5):
  df_as_np=df.to_numpy()
  X=[]
  y=[]
  for i in range(len(df_as_np)-window_size):
    row=[[a] for a in df_as_np[i:i+5]]
    X.append(row)
    label=df_as_np[i+5]
    y.append(label)
  return np.array(X),np.array(y)
```

```python
WINDOW_SIZE=5
X,y=df_to_X_y(temp,WINDOW_SIZE)
X.shape,y.shape
```

```python
X_train,y_train=X[:60000],y[:60000]
X_val,y_val=X[60000:65000],y[60000:65000]
X_test,y_test=X[65000:],y[65000:]
X_train.shape,y_train.shape,X_val.shape,y_val.shape,X_test.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1=Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8,'relu'))
model1.add(Dense(1,'linear'))

model1.summary()
```

```python
cp = ModelCheckpoint('model1/.keras',save_best_only = True)
model1.compile(loss=MeanSquaredError(),optimizer=Adam(learning_rate=0.0001),metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=10,callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model1=load_model('model1/.keras')
```

```python
train_predictions=model1.predict(X_train).flatten()
train_results=pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals':y_train})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][:100])
plt.plot(train_results['Actuals'][:100])
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
rmse1 = np.sqrt(mean_squared_error(train_results['Train Predictions'], train_results['Actuals']))
Maximum_train=(train_results['Actuals']).max()
Mean_train=(train_results['Actuals']).mean()
accuracy_train1=100-(rmse1/Maximum_train)*100
accuracy_train2=100-(rmse1/Mean_train)*100
print(f"LSTM ACCURACY wrt Maximum: {accuracy_train1}")
print(f"LSTM ACCURACY wrt Mean: {accuracy_train2}")
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import r2_score

mae1 = mean_absolute_error(train_results['Actuals'],train_results['Train Predictions'])
mse1 = mean_squared_error(train_results['Actuals'],train_results['Train Predictions'])
mape1 = mean_absolute_percentage_error(train_results['Actuals'],train_results['Train Predictions'])
r21 = r2_score(train_results['Actuals'],train_results['Train Predictions'])
```

```python
print("Mean Absolute Error = ", mae1)
print("Mean Squared Error = ", mse1)
print("Root Mean Squared Error =", rmse1)
print("Mean Absolute Percentage Error =", mape1)
print("r2 Score =", r21)
```

```python
val_predictions=model1.predict(X_val).flatten()
val_results=pd.DataFrame(data={'Val Predictions':val_predictions, 'Actuals':y_val})
val_results
```

```python
import matplotlib.pyplot as plt
plt.plot(val_results['Val Predictions'][:100])
plt.plot(val_results['Actuals'][:100])
```

```python
rmse2 = np.sqrt(mean_squared_error(val_results['Val Predictions'], val_results['Actuals']))
Maximum_val=(val_results['Actuals']).max()
Mean_val=(val_results['Actuals']).mean()
accuracy_val1=100-(rmse1/Maximum_val)*100
accuracy_val2=100-(rmse1/Mean_val)*100
print(f"LSTM ACCURACY wrt Maximum: {accuracy_val1}")
print(f"LSTM ACCURACY wrt Mean: {accuracy_val2}")
```

```python
mae2 = mean_absolute_error(val_results['Actuals'],val_results['Val Predictions'])
mse2 = mean_squared_error(val_results['Actuals'],val_results['Val Predictions'])
mape2 = mean_absolute_percentage_error(val_results['Actuals'],val_results['Val Predictions'])
r22 = r2_score(val_results['Actuals'],val_results['Val Predictions'])
```

```python
print("Mean Absolute Error = ", mae2)
print("Mean Squared Error = ", mse2)
print("Root Mean Squared Error =", rmse2)
print("Mean Absolute Percentage Error =", mape2)
print("r2 Score =", r22)
```

```python
test_predictions=model1.predict(X_test).flatten()
test_results=pd.DataFrame(data={'Test Prediction':test_predictions[1:],'Actuals': y_test[:-1]})
test_results
```

```python
plt.plot(test_results['Test Prediction'][:100])
plt.plot(test_results['Actuals'][:100])
```

```python
rmse3 = np.sqrt(mean_squared_error(test_results['Test Prediction'], test_results['Actuals']))
Maximum_test=(test_results['Actuals']).max()
Mean_test=(test_results['Actuals']).mean()
accuracy_test1=100-(rmse1/Maximum_test)*100
accuracy_test2=100-(rmse1/Mean_test)*100
print(f"LSTM ACCURACY wrt Maximum: {accuracy_test1}")
print(f"LSTM ACCURACY wrt Mean: {accuracy_test2}")
```

```python
mae3 = mean_absolute_error(test_results['Actuals'],test_results['Test Prediction'])
mse3 = mean_squared_error(test_results['Actuals'],test_results['Test Prediction'])
mape3 = mean_absolute_percentage_error(test_results['Actuals'],test_results['Test Prediction'])
r23 = r2_score(test_results['Actuals'],test_results['Test Prediction'])
```

```python
print("Mean Absolute Error = ", mae3)
print("Mean Squared Error = ", mse3)
print("Root Mean Squared Error =", rmse3)
print("Mean Absolute Percentage Error =", mape3)
print("r2 Score =", r23)
```
